### Statement of the Problem
We have data which are event counts binned into a spectrum in calorimetric energy. There is a spectrum generated by monte carlo and decomposed into neutral current event and charged current event components. There is also a data spectrum containing reconstructed events observed in the NOvA near detector. The spectra are created by applying a suite of cuts to files containing reconstructed event information.

Our model therefore is the suite of cuts (each of which has one or several parameters, which we will collectively label $\theta$), along with the proposition that the data and monte carlo spectra are in agreement. The data are the data and Monte Carlo event distributions (because the MC has no free parameters). Therefore, when we calculate $p(D|M,\theta)$, we want to calculate the probability that the data and monte carlo spectra (surviving the model's cuts at $\theta$ parameter values) are in agreement.

### Model
The cuts used are described in NusCuts.h in the root directory of this module. The parameters are the values that can be tuned in each cut, for example:
Hits per plane
lid gap
X, Y, Z start/stop for the fiducial volume
X, Y, Z start/stop for the containment cut
Number of hits
kalman track length (kalman is the module that reconstructed the track)
remid PID (remid is a particular particle ID algorithm)
elecid.ann
numucontpid
[cosmic rejection is not applied at the near detector (ND)]
Vertex position (X,Y,Z)

In addition, our model assumes that the data spectra have bin contents Poisson distributed with mean and variance given by the bin content of the total MC spectrum. This is well motivated because the data are the result of a Poisson process, and if the MC models the data well the mean in a given bin for the data will be the bin content for the MC. To a certain extent, this takes into account reconstruction efficiency, because the monte carlo are subjected to the same reconstruction pipeline as the data (of course, in MC we know whether the reconstruction was successful). Monte Carlo are normalized to the number of protons on target in the data file (that is, the flux of protons producing neutrinos during the time the data was taken, which means the MC and data should be seeing the same event rate)

### Likelihood Function
The likelihood function is given by a Poisson distribution, multiplying a term for each bin. The binning is somewhat poorly motivated, I will admit.
$$ P(D|M,\theta)=\Pi_{bins}\frac{\lambda_i^{D_i}e^{-\lambda_i}}{D_i!} $$
Where $\lambda_i$ is the value of the monte carlo spectrum in the ith bin, and $D_i$ is the value of the data spectrum in the $i$th bin.

We want to maximize this likelihood subject to a constraint on the figure of merit, which in this experiment is
$$FOM=\frac{signal}{\sqrt{signal+background}}$$
I suggest a threshold which is $75\%$ of the current FOM, but this is somewhat arbitrary and it would probably be worthwhile to have a more objective way of optimizing for FOM and likelihood simultaneously.
